```python
import numpy as np
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
%matplotlib inline

from keras.layers import Dense, InputLayer
from keras.models import Sequential

from sklearn.model_selection import train_test_split
```

```python
data_1 = pd.read_csv('train.csv')
```

```python
data_1.head()
```

```python
data_2 = pd.read_csv('test.csv')
```

```python
data_2.head()
```

```python
# create random number generator
seed = 42
rng = np.random.RandomState(seed)
```

```python
# distribution of classes in the dataset
data_1['class'].value_counts()
```

```python
# load images and store it in numpy array

# empty list to store the images
X = []
# iterating over each image
for img_name in data_1.image_names:
    # loading the image using its name
    img = plt.imread('images/' + img_name)
    # saving each image in the list
    X.append(img)
    
# converting the list of images into array
X=np.array(X)

# storing the target variable in separate variable
y = data_1['class'].values
#y = data_1.class.values
```

```python
# shape of the image
X.shape
```

```python
# visualizing images
fig,ax = plt.subplots(nrows=1,ncols=5,figsize=(20,20))
for i in range(5):
    ax[i].imshow(X[i*400])
    ax[i].axis('off')
```

```python
# converting 3 dimensional image to 1 dimensional image(flattening)
X = X.reshape(X.shape[0], 218*178*3)
X.shape
```

```python
# minimum and maximum pixel values of images
X.min(), X.max()
```

```python
# creating a training and validation set
X_train, X_valid, y_train, y_valid=train_test_split(X,y,test_size=0.3, random_state=seed)
```

```python
# shape of training and validation set
(X_train.shape, y_train.shape), (X_valid.shape, y_valid.shape)
```

```python
# defining the architecture of the model
model=Sequential()
model.add(InputLayer(shape=(X_train.shape[1],)))
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# compiling the model
model.compile(loss='binary_crossentropy', optimizer="sgd", metrics=['accuracy'])
```

```python
# training the model
model_history = model.fit(X_train, y_train, epochs=50, batch_size=128,validation_data=(X_valid,y_valid))
```

```python
# Get prediction probabilities
prediction_probabilities = model.predict(X_valid)[:, 0]  # Get probabilities

# Apply threshold to get class predictions
threshold = 0.5
predictions = (prediction_probabilities > threshold).astype(int)  # Convert probabilities to class labels
```

```python
# accuracy on validation set
from sklearn.metrics import accuracy_score
print('Accuracy on validation set:', accuracy_score(y_valid, predictions), '%')
```

```python
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
# pull out the original images from the data which correspond to the validation data
_, valid_Gender, _, valid_y = train_test_split(data_1.image_names.values, y, test_size=0.3, random_state=seed)

# get a random index to plot image randomly
index = rng.choice(range(len(valid_Gender)))

# get the corresponding image name and probability
img_name = valid_Gender[index]
prob = (prediction_probabilities * 100).astype(int)[index]

# read the image
img = plt.imread('images/' + img_name)

# print probability and actual class
print('Model is', prob , '% sure that it is female')
print('And actual class is ', valid_y[index])

# plot image
plt.imshow(img)
```

```python
from keras.optimizers import Adam

# Define the model with ReLU in the hidden layer
model = Sequential([
    InputLayer(shape=(X_train.shape[1],)),  # Using 'shape' instead of 'input_shape'
    Dense(100, activation='relu'),  # Hidden layer with ReLU activation
    Dense(1, activation='sigmoid')  # Output layer with sigmoid activation for binary classification
])

# Compile the model with Adam optimizer
model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
```

```python
# summary of the model
model.summary()
```

```python
# training the model for 50 epochs
model_history = model.fit(X_train, y_train, epochs=50, batch_size=128,validation_data=(X_valid,y_valid))
```

```python
from sklearn.metrics import accuracy_score

# Predict probabilities and convert to binary class predictions
# A common threshold for binary classification is 0.5
threshold = 0.5
prediction_probabilities = model.predict(X_valid)  # Predict probabilities for validation set

# Apply threshold to determine class labels (binary classification)
predictions = (prediction_probabilities > threshold).astype(int)

# Calculate accuracy
accuracy = accuracy_score(y_valid, predictions)  # Compute accuracy score

print('Accuracy on validation set:', accuracy * 100, '%')  # Print accuracy as a percentage
```

```python
# Train the model
model_history = model.fit(
    X_train, 
    y_train, 
    epochs=100,  # Number of epochs
    batch_size=128,  # Batch size
    validation_data=(X_valid, y_valid)  # Validation data
)
```

```python
from sklearn.metrics import accuracy_score

# Get prediction probabilities for the validation set
prediction_probabilities = model.predict(X_valid)

# Convert probabilities to class predictions using a threshold (0.5 for binary classification)
threshold = 0.5
predictions = (prediction_probabilities > threshold).astype(int)

# Calculate accuracy
accuracy = accuracy_score(y_valid, predictions)

# Display accuracy as a percentage
print(f'Accuracy on validation set: {accuracy * 100:.2f}%')
```
